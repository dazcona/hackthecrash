```python
import csv
import pandas as pd
import os
```

```python
# 1. Group all vehicles by accident_id in a dict{accident_id:[array with all vehicles]}

vehicles = {}

with open("../../data/vehicles.csv") as f:
    
    lines = f.readlines()

    for line in lines[1:]:
    
        csv_row = line.split(',')
        accident_id = int(csv_row[0])
        vehicles.setdefault(accident_id, [])
        vehicles[accident_id].append(csv_row[1:])
```

```python
'Number of vehicles: {:,}'.format(len(vehicles))
```

```python
# 2. Write all vehicles info in a CSV file grouped by accident_id

with open("../../data/text.csv", 'w') as csvfile:
    
    writer = csv.writer(csvfile, delimiter=',', quoting=csv.QUOTE_MINIMAL)
    writer.writerow(['accident_id', 'all_vehicles'])

    for accident_id, info in vehicles.items():
        
        # print(len(info))
        # print(accident_id)
        
        all_vehicles = ''
        for vehicle in info:
            # import pdb; pdb.set_trace()
            all_vehicles += ' ' + ' '.join(vehicle)
        
        writer.writerow( [accident_id, all_vehicles] )
```
